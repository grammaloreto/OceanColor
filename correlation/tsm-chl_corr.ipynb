```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dfa = pd.read_csv("chl-tsm-19-01.csv")
dfb = pd.read_csv('chl-tsm-26-01.csv')
dfc = pd.read_csv("chl-tsm-21-02.csv")
dfd = pd.read_csv("chl-tsm-21-02-II.csv")
dfe = pd.read_csv("chl-tsm-22-02.csv")
```

## 19-01

```python
xa = dfa['tsm']
ya = dfa['chl']
```

```python
scipy.stats.pearsonr(xa, ya) # Pearson s r
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(dfa.corr(), annot=True, cmap='coolwarm')
```

```python
#Scatter plot
plt.scatter(dfa["tsm"],dfa["chl"])
dfa['color']= np.where( value==True , "#9b59b6", "#3498db")
sns.regplot(data=dfa, x="tsm", y="chl", fit_reg=False, scatter_kws={'facecolors':dfa['color']})

plt.xlabel('tsm')
plt.ylabel('chl')
```

## All days Chlorophyll

```python
df_chl = pd.concat([ya, yb, yc, yd, ye], axis=1)
```

```python
df_chl.describe()
```

Correlation all chlorophylls 

```python
plt.figure(figsize=(4,2))
sns.heatmap(df_chl.corr(), annot=True, cmap='coolwarm')
```

```python
m_chl = df_chl.mean()
```

```python
# spatial chl average
(markers, stemlines, baseline) = plt.stem(m_chl)
plt.setp(markers, marker='D', markersize=8, markeredgecolor="orange", markeredgewidth=2)
plt.ylabel('Chl (mg/m3)', {'fontsize': 15})
plt.show()
```

### All days TSM
