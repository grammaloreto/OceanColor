```python
import scipy.stats
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dfa = pd.read_csv("chl-tsm-19-01.csv")
dfb = pd.read_csv('chl-tsm-26-01.csv')
dfc = pd.read_csv("chl-tsm-21-02.csv")
dfd = pd.read_csv("chl-tsm-21-02-II.csv")
dfe = pd.read_csv("chl-tsm-22-02.csv")
```

## 19-01

```python
xa = dfa['tsm']
ya = dfa['chl']
```

```python
scipy.stats.pearsonr(xa, ya) # Pearson s r
```

```python
plt.figure(figsize=(4,2))
sns.heatmap(dfa.corr(), annot=True, cmap='coolwarm')
```

```python
plt.scatter(dfa["tsm"],dfa["chl"])
plt.xlabel('tsm')
plt.ylabel('chl')
```

```python
mean1 = ya.mean()
mean1
```
