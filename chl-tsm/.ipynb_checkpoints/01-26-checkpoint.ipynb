```python
%matplotlib inline
import netCDF4 
import matplotlib.pyplot as plt
import numpy as np
import xarray as xr
import pandas as pd
```

Location: Golfo de Penas (chilean Patagonia).

Geo-cordinates (image subset):
- NLb = -44.531
- WLb = -75.854
- SLb = -47.584
- ELb = -74.462

DATE: 2022/01/26


```python
# Xarray dataset from SNAP (Sentinel Application Plataform) C2RCC Processor
file = 'subset_penas20220126T135830_C2RCC.nc'
ds = xr.open_dataset(file)
print('data size:',ds.nbytes/1e6, 'MB')
ds
```

```python
# Checking all variables
nc = netCDF4.Dataset(file, mode='r')
nc.variables.keys()
```

```python
# Variables of interest
con_chl = nc.variables['conc_chl']
con_tsm = nc.variables['conc_tsm']
```

```python
chl = np.array(con_chl)
tsm = np.array(con_tsm)
```

```python
# DataFrame with chl and tsm values 
df = pd.DataFrame(data={'chl' : chl.flatten(), 'tsm' : tsm.flatten()})
```

```python
df.describe()
```

```python
# uncertain or error of chl and tsm values
u_chl = nc.variables['unc_chl']
u_tsm = nc.variables['unc_tsm']
```

```python
chl_unc = np.array(u_chl)
tsm_unc = np.array(u_tsm)
```

```python
# Dataframe
df_u = pd.DataFrame(data={'chl_unc' : chl_unc.flatten(), 'tsm_unc' : tsm_unc.flatten()})
```

```python
df_u.describe()
```

### Chlorophyll a

```python
# quick view of values distribution... most values between 0 and 20
plt.hist(chl)
```

```python
# chl concentration under 5 mg/m-3
chl_values = df.chl[df["chl"]<=5.00]
```

```python
chl_values.describe()
```

```python
# chl concentration over 5 mg/m-3
chl_high_values = df.chl[df["chl"]>=20.00]
```

```python
chl_high_values.describe()
```

```python
# Chlorophyll a map with matplotlib
ds.conc_chl.plot(vmin = 0, vmax = 20)
```

```python
# another map
fig,ax=plt.subplots(figsize=(12,6))
ds.conc_chl.plot(vmin=0, vmax=20, cmap='jet',ax=ax)
```

### SNAP picture chlorophyll a concentration

```python
#Log10 scale of chlorophyll a map

import rasterio
import cartopy.crs as ccrs
proj = ccrs.Mercator()
data_crs = ccrs.PlateCarree()

plt.figure( figsize=(10,10))
ax = plt.axes(projection=proj)


# defining the variables to plot
## log transformation
conc_chl = np.log10(ds.conc_chl)
lat = ds.lat
lon = ds.lon

plt.pcolor(lon, lat, conc_chl, transform = data_crs)

plt.colorbar(label = 'Chlorophyll a', shrink=0.5)

ax.coastlines('10m')

plt.show()
```

###### chlorophyll uncertainity

```python
plt.hist(chl_unc)
```

```python
# unc_chlorophyll a map. It is obvious, if concentration is higher the unc too.
ds.unc_chl.plot(vmin = 0, vmax = 5)
```

This case can be considered as an Optical 2 waters which are coastal waters that are influenced by terrestrial run-off, and thus are optically also influenced by total suspended matter (TSM) and by CDOM (colored dissolved organic matter).

However are also strongly influenced by TSM, especially waters that have a strong tidal influence where the resuspension of sediments may reach much further off-shore
